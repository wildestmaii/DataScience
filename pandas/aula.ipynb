```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empresas = [
    'Apple', 'Google', 'Amazon',
    'Microsoft', 'Tencent',
    "McDonald's", 'Visa',
    'Facebook', 'Alibaba',
    'Louis Vuitton']
```

```python
bilhoes = [
    947.062, 819.573, 705.646,
    611.460, 214.023, 196.526,
    191.032, 186.421, 169.966,
    124.273
]
```

```python
plt.plot(empresas, bilhoes, color = 'skyblue', marker='o')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhoes', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhoes de dolares'], loc = 'right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (14,6)
plt.bar(empresas, bilhoes, color='skyblue')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhoes', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhoes de dolares'], loc = 'right', fontsize=15)
plt.show()
```

```python
cores = [
    'black', 'dimgray', 'dimgrey', 'gray', 'grey',
    'darkgray', 'darkgrey', 'silver', 'lightgray', 'lightgrey'
]
```

```python
plt.rcParams["figure.figsize"] = (15,8)
plt.bar(empresas, bilhoes, color=cores, edgecolor='black')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhoes', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhoes de dolares'], loc = 'right', fontsize=15)
plt.show()
```

```python
plt.style.use('seaborn-white')
plt.rcParams["figure.figsize"] = (15,8)
plt.barh(empresas, bilhoes, color=cores, edgecolor='black')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhoes', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhoes de dolares'], loc = 'right', fontsize=15)
plt.show()
```
